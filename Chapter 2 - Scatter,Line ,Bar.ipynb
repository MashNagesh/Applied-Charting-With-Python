```python
plt.figure()
# plot a data series 'Tall students' in red using the first two elements of x and y
plt.scatter(x[:2], y[:2], s=100, c='red', label='Tall students')
# plot a second data series 'Short students' in blue using the last three elements of x and y 
plt.scatter(x[2:], y[2:], s=100, c='blue', label='Short students')
plt.xlabel(" the x axis")
plt.ylabel("the y")
plt.title("trial charts")
plt.legend(loc = 4,frameon= False)
```

```python
#Line Plot trials
import numpy as np
linear = np.array([1,2,3,4,5,6])
quad = linear **2

plt.Figure()
plt.plot(linear,'-o',quad,'-o')
plt.xlabel(" Index as axis ")
plt.ylabel(" Linear and Quadratic Values ")
plt.title (" Line Plot Trial ")
plt.legend([" Linear", " Quadratic"])
```

```python
plt.gca().fill_between (range(len(linear)),linear,quad,facecolor ='green', alpha =0.25)
```

```python
import pandas as pd
plt.figure()
observation_dates = np.arange('2017-01-01','2017-01-07',dtype='datetime64[D]')
observation_dates = list(map(pd.to_datetime,observation_dates))
print(observation_dates)
plt.plot(observation_dates,linear,'-o', observation_dates,quad,'-o')
plt.legend(["linear","Quad"])
```

```python
#Bar Charts
plt.figure()
xvals = range(len(linear))
plt.bar(xvals,quad,width=0.3,color='teal')
```

```python
new_xvals = []
for i in xvals:
    new_xvals.append(i-0.3)
plt.bar(new_xvals,linear,color='cyan',width = 0.3)
```

```python
#Stacked bar
plt.figure()
plt.bar(xvals,linear,color='blue',width = 0.3)
plt.bar(xvals,quad,color ='olive',width = 0.3,bottom=linear)
```

```python
#Creating horizontal bar graphs
plt.figure()
plt.barh(xvals,linear,height =0.5,color ='maroon')
plt.barh(xvals,quad,height =0.5,left = linear,color='brown')
```
